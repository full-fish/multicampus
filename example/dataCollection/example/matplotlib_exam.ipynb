```python
data = {
    '공부시간': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    '성적': [40, 45, 50, 55, 60, 70, 75, 80, 85]
}

df = pd.DataFrame(data)

df.plot(x="공부시간", y="성적", kind="scatter", color="purple")
plt.title("공부시간과 성적의 관계")
plt.xlabel("공부시간(시간)")
plt.ylabel("성적(점수)")
plt.grid(True)
plt.show()
```

```python
data = {
    '월': ['1월', '2월', '3월', '4월', '5월'],
    '매출': [100, 150, 130, 170, 160],
    '비용': [70, 90, 80, 120, 110]
}

df = pd.DataFrame(data)

df.plot(x="월", y = ['매출', '비용'], marker='o')
plt.title("월별 매출 및 비용 추이")
plt.ylabel("금액(만원)")
plt.grid(True)
plt.show()
```

```python
data = {
    '부서': ['영업', '개발', '디자인', '인사', '마케팅'],
    '인원수': [25, 40, 15, 10, 20],
    '평균연봉': [5200, 6800, 4700, 5000, 5600]
}
df = pd.DataFrame(data)

ax = df.plot(x="부서", y="인원수", kind="bar", color="skyblue", figsize=(8,4))

ax2 = ax.twinx()

# df['평균연봉'].plot(ax=ax2, color='orange', marker='o', linewidth=1, label="평균연봉")

ax.set_ylabel("인원수")
ax2.set_ylabel("평균연봉(만원)")
ax.set_title("부서별 인원 및 평균연봉 비교")

bar, label1 = ax.get_legend_handles_labels()
line, label2 = ax2.get_legend_handles_labels()
ax.legend(bar + line, label1+label2, loc ="upper right")

# plt.legend(loc="upper right")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
fig, axes = plt.subplots(2, 2, figsize=(8, 6)) # 도화지 객체를 만들고 그안에 소속한 그래프객체를 생성해서 리턴

axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title("sin(x)")

axes[0, 1].plot(x, np.cos(x))
axes[0, 1].set_title("cos(x)")

axes[1, 0].plot(x, np.tan(x))
axes[1, 0].set_title("tan(x)")

axes[1, 1].hist(np.random.randn(1000))
axes[1, 1].set_title("Histogram")

plt.tight_layout()
plt.show()
```

```python
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y1, color="blue")
plt.title("sin(x)")

plt.subplot(1, 2, 2)
plt.plot(x, y2, color="red")
plt.title("cos(x)")

plt.tight_layout()
plt.show()
```
